```python
# Import Required Libraries
#import pandas as pd
import numpy as np
```

```python
#1 Load dataset
#Dataset is on my public github repository, 
#don't need to download it, just use the link to access it

url = "https://raw.githubusercontent.com/AUunkle/ML_in_data/main/hw2/iris.data"
df = np.genfromtxt(url, delimiter=",",dtype=float, usecols=[0, 1, 2, 3])
```

```python
#2 Check the shape of the dataset
if df.ndim != 2:
    df = df.reshape(-1, 4)
  
```

```python
 #3 Calculate mean, median, standard deviation for the 1-st column of the dataset
meanVal = np.mean(df[:, 0])
medianVal = np.median(df[:, 0])
stdVal = np.std(df[:, 0])

print(meanVal)
print(medianVal)
print(stdVal)
```

```python
#4 change to NaN  20 random elements by unique index
nanInd = np.random.choice(df.size, 20, replace=False)

df[np.unravel_index(nanInd, df.shape)] = np.nan
print(df)
```

```python
#5 define pos's  of nan in 1-st column

nanInd_1_clmn = np.where(np.isnan(df[:, 0]))
print(nanInd_1_clmn)
```

```python
#6 filtered array by statement: value > 1.5  in the 3-rd clmn, value  < 5.0 in the 1-st clmn (зберегти у іншу змінну)

flt_df = df[(df[:, 2] > 1.5 ) & (df[:, 0] < 5.0)]

print( flt_df)
```

```python
#7 change nan to 0.0
df = np.nan_to_num(df, nan=0.0)
print (df)
```

```python
#8 count all unique values
uniqueVal, counts = np.unique(df, return_counts=True)
dictUniqVal = dict(zip(uniqueVal, counts))
np.set_printoptions(legacy="1.25")
print(dictUniqVal)

```

```python
#9 split the dataset into 2 equal parts by rows
df1, df2 = np.array_split(df, 2, axis=0)
print(df1)
print("-------------------")
print(df2)
```

```python
#10 sort by the 1-st column df1 acsending, df2 descending orders
df1 = df1[np.argsort(df1[:, 0])]
df2 = df2[np.argsort(df2[:, 0])[::-1]]
print(df1)
print("-------------------")
print(df2)
```

```python
#11 concatenate df1 and df2
dfCnct = np.concatenate((df1, df2))
print(dfCnct)
```

```python
#12 find the most repetable element df_concatenated
uniqueVal, counts = np.unique(dfCnct, return_counts=True)
max_count = np.max(counts)
mostRpt = uniqueVal[counts == max_count]

print(mostRpt)
```

```python
#13 function that multiply by 2 all el less then mean and divide by 4 above the mean separatly for each column
def change2by4Rule(_df, _clmn=-1):
    means = np.mean(_df, axis=0)

    if _clmn == -1:
        for i in range(_df.shape[1]):
            _df[:,i] = np.where(_df[:,i] < means[i], _df[:,i] * 2, _df[:,i] / 4)
    else :
        _df[:, _clmn] = np.where(_df[:, _clmn] < means[_clmn], _df[:, _clmn] * 2, _df[:, _clmn] / 4)
        
    return _df
```

```python
#14 apply function for the 3-rd column of the dataset
print(dfCnct)
change2by4Rule(dfCnct,2)
print("-------------------")
print(dfCnct)
```
